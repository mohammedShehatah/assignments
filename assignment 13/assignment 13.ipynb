```python
import requests
from bs4 import BeautifulSoup
```

```python
def books_data (site):
    feed_back = requests.get(site)
    src_page = BeautifulSoup(feed_back.text,"html.parser")
    uls = src_page.find_all("ul")
    lis = uls[2].find_all("a")
    
    links = []
    cat_names = []
    
    for link in lis :
        links.append("http://books.toscrape.com/"+link.get("href"))
        cat_names.append(link.get_text().strip())
        
    
    catigories = {}
    
    i = 0
    for cat in cat_names :
        catigories[cat] = requests.get(links[0])
        i = i+1
        
    with open ("books_data.text" , "a+", encoding="utf-8") as f :
        
        for key in catigories :
            catigories[key] = BeautifulSoup (catigories[key].text , "html.parser")
            articls = catigories[key].find_all("article",attrs={"class":"product_pod"})
            print ("******************\nnew catigory\n******************\n")
            f.write("******************\nnew catigory\n******************\n")
            
            for article in articls :
                book_name = article.find("h3").get_text()
                price = float( article.find("p", attrs= {"class":"price_color"}).get_text().split('£')[1] )
                rate = article.find("p").get("class")[1]
                f.write (f"book name : {book_name}\nprice : {price}\nrate : {rate}\n---------------------\n")
                print(f"book name : {book_name}\nprice : {price}\nrate : {rate}\n---------------------\n")
                             
                             
                        

        
    
    
    
    
```

```python
books_data ("http://books.toscrape.com/index.html")
```
